## OpenCV

```python
import cv2   # import OpenCV
print("OpenCV version", cv2.__version__)
```

```python
#load an image
img = cv2.imread('assets/0922-1.tiff')
```

```python
# display an image
cv2.imshow('My Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
print('dimensions = ', img.shape)
```

```python
img_new = cv2.resize(img, (250,400))
cv2.imshow('My Image', img_new)
cv2.waitKey(0); cv2.destroyAllWindows(); cv2.waitKey(1)
```

```python
img_new = cv2.resize(img, (0,0), fx=1.5, fy=1.5)
cv2.imshow('My Image', img_new)
cv2.waitKey(0); cv2.destroyAllWindows(); cv2.waitKey(1)
```

```python
img_new = cv2.rotate(img, cv2.ROTATE_180)
cv2.imshow('My Image', img_new)
cv2.waitKey(0); cv2.destroyAllWindows(); cv2.waitKey(1)
```

```python
cv2.imwrite('assets/img_rotate.jpg', img_new)
```
